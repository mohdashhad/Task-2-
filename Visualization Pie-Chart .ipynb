```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
excel_path = (r"C:\Users\Hp\Downloads\Dubai_Final_Keyword_Report.xlsx") # Use your file path here
df = pd.read_excel(excel_path, sheet_name='Sheet1')

# Count the number of occurrences of each source URL
source_counts = df['Source URL'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(source_counts.values, labels=source_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Count of Source URL')
plt.axis('equal')  # Ensures the pie chart is a circle
plt.tight_layout()
plt.show()
```
